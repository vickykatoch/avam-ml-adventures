```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('../data/basic/housing-data.csv',delimiter=',',skip_header=1);
x = data[:,0]
y = data[:,1]
m = len(x)
```

```python
x_mean = x.mean()
y_mean = y.mean()
print(f'Mean [X,Y] : [{x_mean}, {y_mean}]')
x_sd = x.std()
y_sd = y.std()
print(f'STD Dev [X,Y] : [{x_sd}, {y_sd}]')
x_zscores = (x_mean - x)/x_sd
y_zscores = (y_mean - y)/y_sd
cor_coff = (x_zscores * y_zscores).sum()/m
```

```python
slope_m = cor_coff * y_sd/x_sd
intercept_b = y_mean - slope_m * x_mean
predicted = slope_m * x + intercept_b
```

```python
output = np.hstack([data,predicted.reshape(m,1)])
```

```python
plt.scatter(x, y)
plt.xlabel('Area')
plt.ylabel('Cost')
plt.title(f'{slope_m}.x + {intercept_b}')
plt.plot(x, predicted, color='g');
```
